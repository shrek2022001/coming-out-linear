```python
df['CRIM'].fillna(df['CRIM'].median(),inplace=True) 
```

```python
df['CRIM'].isnull().sum()
```

```python

df['ZN'].fillna(df['ZN'].mean(),inplace=True)
df['INDUS'].fillna(df['INDUS'].mean(),inplace=True)
df['CHAS'].fillna(df['CHAS'].mean(),inplace=True)
df['LSTAT'].fillna(df['LSTAT'].mean(),inplace=True)
df['AGE'].fillna(df['AGE'].mean(),inplace=True)
```

```python
df.isnull().sum().sum() #removed all the null values 
```

```python
from scipy import stats     #correlation of two variable
stats.pearsonr(df['RM'],df['MEDV'])
```

```python
X=df.iloc[:,0:13]
X
```

```python
y=df.iloc[:,13]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
print("The dataframe dimension is :",df.shape)
print("The X_train dimension is :",X_train.shape)
print("The X_test dimension is :",X_test.shape)
print("The y_train dimension is :",y_train.shape)
print("The y_test dimension is :",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
PRED=model.predict([df.iloc[5,0:13]])
```

```python
print(PRED)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
y_pred_linear_train = model.predict(X_train)
r2_score_linear_train = r2_score(y_train, y_pred_linear_train)
```

```python
print(r2_score_linear_train)
```

```python
y_pred_linear_test = model.predict(X_test)
r2_score_linear_test = r2_score(y_test, y_pred_linear_test)
```

```python
r2_score_linear_test
```

```python
rmse_linear = (np.sqrt(mean_squared_error(y_test, y_pred_linear_test)))
print(rmse_linear)
```
